```python
student_score['Exams'] = np.random.randint(30,65,40). reshape(40,1)#creating a new column at the end of the dataframe
student_score
```

```python
student_score['TotalScore']=student_score['CA1'] + student_score['CA2'] + student_score['CA3'] + student_score['Exams']#creating a new column from existing column

student_score
```

```python
student_score.loc['Empty'] = 0 #creating a new row
student_score
```

```python
student_score.drop('TotalScore',axis = 1,inplace =True)
student_score.drop('Empty',axis =0,inplace =True)
student_score
#dropping rows and columns, where axis set 0 means what you want to drop is located on the row axis while axis set to 1 means what you want to drop is located on the column axis
```
